```python
import sys
import os
from pathlib import Path
from dotenv import load_dotenv

# 1. Cargar variables del .env
load_dotenv()

# 2. Obtener la ruta absoluta a la raíz del proyecto (carpeta mi_bot)
ROOT = Path(os.getcwd()).resolve().parent   # equivalente a ../ desde /notebooks

# 3. Añadir la raíz al sys.path para permitir importaciones del paquete bot/
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

print("ROOT:", ROOT)

# 4. Construir ruta absoluta del PROMPTS_DB_PATH usando la raíz del proyecto
raw_prompts_path = os.getenv("PROMPTS_DB_PATH", "db/prompts.db")
PROMPTS_DB_PATH = (ROOT / raw_prompts_path).resolve()

print("PROMPTS_DB_PATH:", PROMPTS_DB_PATH)
print("EXISTE prompts.db?:", PROMPTS_DB_PATH.exists())
```
